### If you want to read some specific rows of the dataset use nrows parameters

```python
df = pd.read_csv('D:\Python codes\Pandas\Fortune_10.csv', nrows = 1)
df
```

```python
df = pd.read_csv('D:\Python codes\Pandas\Fortune_10.csv', nrows = 5)
df
```

### If you want to read some specific columns of the dataset use usecols parameters

```python
df = pd.read_csv('D:\Python codes\Pandas\Fortune_10.csv', usecols = [0, 1])
df
```

```python
df = pd.read_csv('D:\Python codes\Pandas\Fortune_10.csv', usecols = [2, 3, 5])
df
```

### If you want to skip rows of the dataset use skiprows parameter

```python
df = pd.read_csv('D:\Python codes\Pandas\Fortune_10.csv', skiprows = [2])
df
```

```python
df = pd.read_csv('D:\Python codes\Pandas\Fortune_10.csv', skiprows = [2, 3, 5])
df
```

### If you want to change index of the dataset

```python
df = pd.read_csv('D:\Python codes\Pandas\Fortune_10.csv', index_col = 'ID')
df
```

```python
df = pd.read_csv('D:\Python codes\Pandas\Fortune_10.csv', index_col = 'Name')
df
```

```python
#changing the name of headers.
df = pd.read_csv('D:\Python codes\Pandas\Fortune_10.csv', header = None, names = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'])
df
```

```python
df = pd.read_csv('D:\Python codes\Pandas\Fortune_10.csv', header = None, prefix = 'Data')
df
```

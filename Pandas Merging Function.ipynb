```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'ID': [1, 2, 3, 4, 5], 'Class': [6, 7, 8, 9, 10]})
df1
```

```python
df2 = pd.DataFrame({'ID': [1, 2, 3, 4, 5], 'Name': ['A', 'B', 'C', 'D', 'E']})
df2
```

```python
pd.merge(df1, df2, on = 'ID')
```

```python
pd.merge(df2, df1, on = 'ID')
```

```python
df3 = pd.DataFrame({'ID': [1, 2, 3, 4, 5], 'Class': [6, 7, 8, 9, 10]})
df3
```

```python
df4 = pd.DataFrame({'ID': [1, 2, 3, 4, 6], 'Class': ['A', 'B', 'C', 'D', 'E']})
df4
```

```python
pd.merge(df3, df4, on = 'ID', how = 'left')
```

```python
pd.merge(df3, df4, on = 'ID', how = 'right')
```

```python
pd.merge(df3, df4, on = 'ID', how = 'outer', indicator = 'Ture')
```

```python
df5 = pd.DataFrame({'ID': [1, 2, 3, 4, 5], 'Class': [6, 7, 8, 9, 10]})
df5
```

```python
df6 = pd.DataFrame({'ID': [6, 7, 8, 9, 10], 'Name': ['A', 'B', 'C', 'D', 'E']})
df6
```

```python
pd.merge(df5, df6, left_index = True, right_index = True)
```
